## 1. Problem Definition 

Given the 5 year daily stock history of Apple (AAPL),can we predict the percentage in stock price from month-to-month

## 2. Data

Our target variable for this project is "Monthly Percentage Change" (MPC). 

__5-year historical stock data on AAPL__
https://www.nasdaq.com/market-activity/stocks/aapl/historical

## 3. Features

- Open Price:
  - price at which stock first traded on the opening of trading day
  - relevant for reflecting investor's sentiment for the day
- High Price:
  - highest price stock traded during the that day
  - indicated high volatility and high buying pressure
- Low Price:
  - lowest price stock traded during that day
  - indicates high volatility and low buying pressure
- Close Price:
  - price at which stock last traded opon closing of trading day
  - benchmark for stock's daily performance
- Volume:
  - total number of shares traded during that trading day
  - indicates interest in stock and/or buying news

```python
# Load Data 
df = pd.read_csv("./5-year-data.csv")
```

```python
# Convert $.. values into ints w/out $ sign

def convert_currency_to_float(value):
    return float(value.replace('$', '').replace(',', ''))

columns_to_convert = ['Close/Last', 'Open', 'High', 'Low']

df[columns_to_convert] = df[columns_to_convert].applymap(convert_currency_to_float)

# Convert Date into num
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df['DayOfYear'] = df['Date'].dt.dayofyear

# Create target var
df['NextDayPctChange'] = df['Close/Last'].pct_change().shift(-1)

df.dropna(inplace=True)


print(df)
```

```python
# Put models into a dict so that we can call a function that will score and test all of them
# Define your models dictionary
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "SVR (Linear Kernel)": SVR(kernel='linear'),
    "SVR (RBF Kernel)": SVR(kernel='rbf'),
    "Random Forest Regressor": RandomForestRegressor(),
    "Gradient Boosting Regressor": GradientBoostingRegressor()
}

# Create a function to fit and score models
def fit_and_score(models, x_train, y_train, x_test, y_test):
    """
    Fits and evaluates given machine learning models. 
    """
    # Set random seed
    np.random.seed(42)
    # Make a dict to keep model scores 
    model_scores = {}
    # Loop through models 
    for name,model in models.items():
        #Fit the model to the data
        model.fit(x_train,y_train)
        #Evaluate the model and append it's score to model_scores
        model_scores[name] = model.score(x_test,y_test)
    return model_scores
```

```python
model_scores = fit_and_score(models=models,
                             x_train=x_train,
                             x_test=x_test,
                             y_train=y_train,
                             y_test=y_test)
model_scores
```

### Hyperparameter turning (by hand)


```python
def tune_model_with_random_search(model, param_distributions, x_train, y_train, n_iter=100, cv=5, verbose=True):
    """
    Perform hyperparameter tuning using RandomizedSearchCV.
    
    Parameters:
    - model: The regression model to tune.
    - param_distributions: The distribution of parameters to sample from.
    - x_train: Training feature dataset.
    - y_train: Training target dataset.
    - n_iter: Number of parameter settings sampled (default=100).
    - cv: Number of folds in cross-validation (default=5).
    - verbose: Verbosity level (default=True).
    
    Returns:
    - The fitted RandomizedSearchCV object.
    """
    np.random.seed(42)
    
    # Setup random hyperparameter search
    rs_model = RandomizedSearchCV(estimator=model,
                                  param_distributions=param_distributions,
                                  n_iter=n_iter,
                                  cv=cv,
                                  verbose=verbose)
    # Fit the model
    rs_model.fit(x_train, y_train)
    
    # Return the fitted model
    return rs_model


# Define the parameter grid for RandomForestRegressor
rf_grid = {
    'n_estimators': randint(10, 200),
    'max_depth': [None, 3, 5, 10],
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20),
}

# Tune RandomForestRegressor
tuned_rf = tune_model_with_random_search(RandomForestRegressor(), rf_grid, x_train, y_train)

# Access the best parameters and score
best_params = tuned_rf.best_params_
best_score = tuned_rf.best_score_

print(f"Best parameters: {best_params}")
print(f"Best score: {best_score}")

```
